```python
#other colleciton of elemets they only conbtain the value.
# With dicitonary you have key value pairs

student_scores = {"Alice":96,"Bob":88,"Charlie":90}
print(student_scores)
print("to get the value")
print("-"*100)
print(student_scores["Bob"])
print(student_scores.get("Bob"))
print("-"*100)
print("Updating the dictionary")
student_scores['TG'] = 55
print(student_scores)
print("deleting a dictionary value")

student_scores.pop("TG")
student_scores.clear()
print(student_scores)

```

```python
student_scores = {"Alice":96,"Bob":88,"Charlie":90}

for key in student_scores.keys():
    print(key)


print("when you want to iterate the values")

for value in student_scores.values():
    print(value)

print("what if you want 2 iterate both key and value")

for key,value in student_scores.items():
  print(f"{key}:{value}")

```
